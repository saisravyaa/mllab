```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('C:\\Users\\sravy\\Desktop\\ml lab\\iris(For PCA Program) (1).csv')
data.head()
```

```python
# Extract the feature columns (excluding the 'species' column)
x = data.drop('species',axis = 1)
x.head()
```

```python
#standardize the data
x_stand = (x - np.mean(x,axis=0))/np.std(x,axis=0)
```

```python
#covariance matrix
cov = np.cov(x_stand,rowvar=False)
```

```python
#calculate egien vectors and eigen values
eigenv,eigenv1 = np.linalg.eig(cov)
```

```python
#sorting
sorted_index = np.argsort(eigenv)[::-1]
eigenv = eigenv[sorted_index]
eigenv1 = eigenv1[:,sorted_index]
```

```python
#plotting
x_pca = np.dot(x_stand,eigenv1[:,:2])
```

```python
# Scatterplot of the first two principal components

plt.figure(figsize=(8, 6))
plt.scatter(x_pca[:, 0], x_pca[:, 1], c=data['species'])
plt.title('Scatterplot of Iris Dataset - First 2 Principal Components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

```
